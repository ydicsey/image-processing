```python
#import the library
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# for cos and sin
import math
```

Create an image g(x,y) whose pixels all have the same gray value of 100. Show the image g(x,y).


```python
M, N = 256, 256
SIGMA = math.sqrt(25)
```

```python
gray = np.full((M, N), 100, dtype=np.int16)
noise = np.full((M, N), 100, dtype=np.int16)
f = np.zeros( [M, N] )
```

```python
for i in range(M-1):
    for j in range(N - 1):
        r = np.random.random()
        phi = np.random.random()
        z_1 = SIGMA * math.cos(2 * math.pi * phi) * math.sqrt(-2 * math.log(r))
        z_2 = SIGMA * math.sin(2 * math.pi * phi) * math.sqrt(-2 * math.log(r))
        noise[i][j] += z_1
        noise[i][j + 1] += z_2

noise = np.clip(noise, 0, 255)
```

```python
f = gray + noise
f = np.clip(f, 0, 255).astype(np.uint8)
```
