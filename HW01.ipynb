Set up ENV

```python
#import the library
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img_path = 'input.jpg'
img = cv.imread(img_path)
img.shape
```

```python
plt.imshow(img)
plt.show()
```

BGR to RGB

```python
image = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

grayscale using OpenCV method

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(gray ,cmap="gray")
plt.show()
```

grayscale using AVG method

```python
gray_AVG = np.sum(image,axis=2)/3
gray_AVG = gray_AVG.astype(int)
plt.imshow(gray_AVG, cmap="gray")
plt.show()
```

compare

```python
f = plt.figure(num=None, figsize=(16, 12), dpi=60)
ax1 = f.add_subplot(2,2, 1)
plt.imshow(gray, cmap="gray")
ax1.title.set_text('OpenCV method')

ax2 = f.add_subplot(2,2, 2)
plt.imshow(gray_AVG, cmap="gray")
ax2.title.set_text('average method')

plt.show()
```

Saving image by opencv

```python
path = "output.jpg"
cv.imwrite(path, gray_AVG)
```
